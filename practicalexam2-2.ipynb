```python
import matplotlib.pyplot as plt
import os

losses = [1.5143, 1.5447, 1.5875, 1.5909] 
heads = [2, 3, 5, 7]

os.makedirs('figures', exist_ok=True)

plt.figure(figsize=(8, 5))
plt.plot(heads, losses, marker='o', linestyle='-', color='b')
plt.title('Validation Loss at Iteration 5000 vs Number of Heads (Layers=5)')
plt.xlabel('Number of Heads')
plt.ylabel('Validation Loss at Iteration 5000')
plt.grid(True)
plt.xticks(heads)
plt.savefig('figures/loss_vs_heads.png')
plt.show()
```

```python
with open('config/train_shakespeare_char.py', 'r') as f:
    config = f.read()

config = config.replace("out_dir = 'out-shakespeare-char'", "out_dir = 'out-code-generation'")
config = config.replace("dataset = 'shakespeare_char'", "dataset = 'code_generation'")

with open('config/train_code_generation.py', 'w') as f:
    f.write(config)

print("Config file successfully created!")
```
